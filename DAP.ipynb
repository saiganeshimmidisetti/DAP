```python
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr)

a = np.zeros((3,3),dtype=int)
print((a))

b=np.ones((2,2),dtype=int)
print(b)

arr_array = np.arange(5)
print(arr_array)

#array manuplation
re_array = arr.reshape(3,2)
print(re_array)

slice_arr = arr[2:5]
print(slice_arr)

#boardcasting
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])
print(a1 + a2)
a3=a1+3
print(a3)

#array manipulation
b=np.array([1,2,3])
c=np.array([4,5,6])
d=np.vstack(b+c)
print(d)

b=np.array([1,2,3])
c=np.array([4,5,6])
d=np.vstack(b+c)
e=np.vstack(b+c)
print(d,e)

#.split()
arr = np.array([1,2,3,4,5,6])
b=np.split(arr,3)
print(b)
```

```python
#dot product
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
c=np.dot(arr1,arr2)
print(c)

#eigen val,vector
d=np.linalg.eig(c)
print(d)
```

```python
#transpose matrix
arr = np.array([[1,2],[3,4]])
q=arr.T
print(q)
```

```python
#.sum(),axis concept
m = np.array([[1,2,3],[4,5,6]])
b=np.sum(m) #add of all elements in matrix
print(b)
print(np.sum(m,axis=0)) #row wice add
print(np.sum(m,axis=1)) #coloumn wice add
```

```python
#statistical ooperations
a=np.array([1,2,3,4,5])
print(np.mean(a))
print(np.median(a))
print(np.var(a))
print(np.std(a))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/citytemp.csv')
at=np.mean(df['AvgTemperature'])
ht=np.max(df['AvgTemperature'])
lt=np.min(df['AvgTemperature'])
threshold=30
days_exceeding_threshold = np.sum(df['AvgTemperature'] > threshold)
print(f'Average Temperature: {at:.2f}C')
print(f'Highest Temperature: {ht}C')
print(f'Lowest Temperature: {lt}C')
print(f'Days Exceeding {threshold}C: {days_exceeding_threshold}')
plt.plot(df['Day'], df['AvgTemperature'])
plt.xlabel('Day')
plt.ylabel('AvgTemperature (C)')
plt.title('Daily Temperatures')
plt.show()
```
